```python
# Using Fake Data 
dataFrame = pd.DataFrame(np.random.randn(100, 4), columns='A B C D'.split())
```

```python
dataFrame.head() 
```

```python
# Scatter Plots 
dataFrame.iplot(kind='scatter', x='A', y='B', mode='markers', size=10)
```

```python
dataFrame.iplot(kind='bar', x='A', y='B', mode='markers', size=10)
```

```python
dataFrame.iplot(kind='box')
```

```python
dataFrame.iplot(kind='bubble', x='A', y='B', size='C')
```
