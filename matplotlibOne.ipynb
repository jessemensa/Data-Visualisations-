```python
import matplotlib.pyplot as plt 
```

```python
%matplotlib inline 
```

```python
import numpy as np 
x = np.linspace(0, 5, 11)
y = x ** 2 
```

```python
x 
```

```python
y 
```

```python
plt.plot(x, y, 'r') # 'r' is the color red 
plt.xlabel('X Axis Title')
plt.ylabel('Y Axis Title')
plt.title('Title of the Visualisation')
plt.show() 
```

```python
# Creating Multiplots on same canvas
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r--') # Color red 
plt.subplot(1, 2, 2) 
plt.plot(x, y, 'g*--'); 
```

```python
# Object Oriented Method 
```

```python
fig = plt.figure() # Create Figure (Empty Canvas) 

# Add set of axes to figure 
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height 
```

```python
# Plot on that set of axes 
axes.plot(x, y, 'b')
```

```python
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure() # Create Figure (Empty Canvas) 

# Add set of axes to figure 
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height 

# Plot on that set of axes 
axes.plot(x, y, 'b')

axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
# Creating a blank Canvas Again 
fig = plt.figure() 

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

# Large Figure Axes 1 
axes1.plot(x, y, 'b')
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('Title One')

# Figure Axes 2 
axes2.plot(x, y, 'r')
axes2.set_xlabel('X Label 2')
axes2.set_ylabel('Y Label 2')
axes2.set_title('Title Two')
```

```python
# Subplots 
fig, axes = plt.subplots() 

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Title')
```

```python
# 1 by 2 subplots 
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
axes # Axes is an array of axes to plot on 
```

```python
for ax in axes: 
    ax.plot(x, y, 'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')

fig 
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes: 
    ax.plot(x, y, 'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')

fig 
plt.tight_layout() 
```

```python
fig = plt.figure(figsize=(8, 4), dpi=100)
```

```python
fig, axes = plt.subplots(figsize=(12, 3))

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title'); 
```

```python
# Legends, Labels and Title 
fig = plt.figure() 

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x**2, label="x**2")
ax.plot(x, x**3, label="x**3")
ax.legend() 
```

```python
# Setting colors, linwidths, linetypes 
fig, ax = plt.subplots() 
ax.plot(x, x**2, 'b.-')
ax.plot(x, x**3, 'g--')
```

```python
fig, ax = plt.subplots() 

ax.plot(x, x+1, color="blue", alpha=0.5)
ax.plot(x, x+2, color="#8B008B")
ax.plot(x, x+3, color="#FF8C00")
```
